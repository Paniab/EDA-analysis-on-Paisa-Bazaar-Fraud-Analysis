# Fraud Risks and Challenges
In financial services, fraud can occur when bad actors try to access loans or credit products using false information or stolen identities. Risks include fake loan applications, misuse of personal data, and unauthorized access to accounts. Paisabazaar faces challenges such as detecting fake identities, monitoring suspicious account activities, and preventing unauthorized product applications.

Here are the main fraud risks for Paisabazaar:

1-Fake loan applications: Fraudsters may use false identities or stolen personal information to apply for loans or credit cards, hoping to get approved and never repay the amount.

2- Identity theft: Criminals can try to impersonate customers or Paisabazaar employees to gain access to sensitive information or financial services.

3- Data misuse: Unauthorized individuals may attempt to access or misuse customer data, leading to privacy breaches or financial loss.

4- Phishing and social engineering: Scammers often send fake emails, SMS, calls, or messages pretending to be Paisabazaar, asking for personal details, banking credentials, or advance payments under the guise of processing applications.

5- Account takeovers: Repeated login attempts from unfamiliar devices or locations may signal fraudsters trying to gain control of customer accounts for malicious purposes.

6- Application inconsistency: Submitting multiple inconsistent or suspicious loan or credit card applications may indicate fraudulent intent.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Step 1: Drop irrelevant or sensitive columns
data.drop(columns=["ID", "Name", "SSN"], inplace =True)

# Step 2: Clean column names (optional but recommended)
data.columns = data.columns.str.strip().str.replace(" ", "_")

# Step 3: Feature Engineering
# Create Debt-to-Income Ratio
data["Debt_to_Income_Ratio"] = data["Outstanding_Debt"] / (data["Annual_Income"] + 1e-5)

# Create Savings Rate
data["Savings_Rate"] = data["Amount_invested_monthly"] / (data["Monthly_Inhand_Salary"] + 1e-5)

# Step 4: (Optional) Filter or group data for analysis
# For example, average delay by occupation
# avg_delay = df.groupby("Occupation")["Delay_from_due_date"].mean()

# Step 5: Check for any obviously wrong values
# For example, negative salaries or ages over 120
data = data[data["Age"] <= 100]
data = data[data["Monthly_Inhand_Salary"] >= 0]

# Step 6: Preview cleaned data
print(data.head())
print("\nData shape:", data.shape)
````

#### Chart - 4 Coorelation Heat Map

````python
# Heat Map visualization code
cols = ["Debt_to_Income_Ratio", "Num_of_Delayed_Payment", "Monthly_Balance", "Credit_Score", "Occupation"]

# Subset the dataset
subset = data[cols].copy()

# Convert categorical variables to numeric using Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
subset["Credit_Score"] = le.fit_transform(subset["Credit_Score"])
subset["Occupation"] = le.fit_transform(subset["Occupation"])

# Compute correlation matrix
corr_matrix = subset.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap for Selected Features", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
````
